```python
links = []

input_file = open('File Directory/GoFundMe_Urls.csv', 'r')

rows = input_file.readlines()
input_file.close()

for i in range(1, len(rows)):
    links.append(rows[i].replace('\n',''))
```

```python
Names=[]
Date=[]
Tags=[]
GoalAmount=[]
AmountRaised=[]
Donors=[]
Shares=[]
Followers=[]
Text=[]
```

```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
path="/Users/mainguyen/Downloads/Purdue Grad/Module 2/590-11-Web Data Analytics/chromedriver"

import time
import random
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
chrome_options = Options()
#chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome('/Users/mainguyen/Downloads/Purdue Grad/Module 2/590-11-Web Data Analytics/chromedriver', options = chrome_options)
time.sleep(3)
driver.quit()

for link_number in range(0,815):
    print(links[link_number])
    
    driver = webdriver.Chrome(path)
    
    driver.get(links[link_number]) # Start the browser and open 'url'
    
    html = driver.page_source # Get the html of the page
    
    driver.quit()

    soup = bs(html, 'html.parser')
    
    #Get Name
    title = soup.find_all('h1', class_='a-campaign-title')
    if len(title) == 0:
        continue
    else:
        Names.append(title[0].text)
    
    #Get amountRaised and amountGoal
    amountTable = soup.find_all('h2', class_= "m-progress-meter-heading")
    amount = amountTable[0].getText()
    AmountRaised.append(amount[:amount.find(' ')])
    GoalAmount.append(amount[amount.find('of')+3:amount.find('g')-1])
    
    #Get donors, shares, followers
    statTable = soup.find_all("span", class_="text-stat-value")
    if len(statTable) != 0:
        Donors.append(statTable[0].getText())
        Shares.append(statTable[1].getText()) 
        Followers.append(statTable[2].getText())
    else:
        Donors.append('0')
        Shares.append('0') 
        Followers.append('0')
        
    #Get dateCreated and tags
    infotable = soup.find_all('div', class_="m-campaign-byline-meta")
    rows = infotable[0].find_all('li')
    Date.append(rows[0].getText())
    Tags.append(rows[1].getText())
    
    #Get description
    description = soup.find_all('div', class_="o-campaign-description")
    if len(description) != 0:
        Text.append(description[0].getText())
    else:
        Text.append('None')
    
    #time.sleep(random.uniform(5,15))
```

```python
for i in range(0,len(Text)-1):
    Text[i] = Text[i].replace('\n','')
```

```python
list = [Names, Date, Tags, GoalAmount, AmountRaised, Donors, Shares, Followers, Text]

for item in list: 
    print(item[len(item)-1])
```

```python
import pandas as pd
results = pd.DataFrame({'Title': Names, 'Date': Date, 'Tags': Tags,
                        'Goal Amount': GoalAmount, 'Amount Raised': AmountRaised,
                        'Donors': Donors, 'Shares': Shares, 'Followers': Followers,
                        'Description': Text})
results.to_csv('GoFundMe.csv')
```
